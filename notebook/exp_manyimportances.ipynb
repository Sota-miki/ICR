```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xgboost as xgb
import xgbfir 
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import log_loss
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import shap
import itertools
```

```python
dataset_df = pd.read_csv(r"C:\Users\ghibl\ICR\data\input\train.csv")
dataset_df.drop("Id",axis = 1, inplace = True)
```

```python
le = LabelEncoder()
dataset_df["EJ"] = le.fit_transform(dataset_df["EJ"])
```

```python
X_df = dataset_df.drop("Class",axis =1)
y_df = dataset_df["Class"].copy()
skf = StratifiedKFold(n_splits = 4, shuffle = True, random_state = 42)
scores = []
```

```python
X_df.head()
```

```python
columns = list(X_df.columns)
com_columns = list(itertools.combinations(columns, 2))
for com_col in com_columns:
    X_df[f"{com_col[0]}_{com_col[1]}_mul"] = X_df[f"{com_col[0]}"] *  X_df[f"{com_col[1]}"]
    X_df[f"{com_col[0]}_{com_col[1]}_sum"] = X_df[f"{com_col[0]}"] +  X_df[f"{com_col[1]}"]
```

```python
X_df.head()
```

```python
evals_results = []
fscore_dicts = []
for tr_idx, va_idx in skf.split(X_df, y_df):
    X_train, X_valid = X_df.iloc[tr_idx], X_df.iloc[va_idx]
    y_train, y_valid = y_df.iloc[tr_idx], y_df.iloc[va_idx]

    weights = [5 if i == 1 else 1 for i in y_train]
    dtrain = xgb.DMatrix(X_train, label = y_train, weight = weights)
    dvalid = xgb.DMatrix(X_valid, label = y_valid)

    evals_result = {}
    params = {'objective': 'binary:logistic','verbosity': 1, 'random_state': 42, 'learning_rate' : 0.1}
    num_round = 500

    watchlist = [(dtrain, 'train'),(dvalid, 'eval')]
    model = xgb.train(params, dtrain, num_round, evals = watchlist, 
                      early_stopping_rounds = 20, evals_result = evals_result)

    evals_results.append(evals_result)
    y_pred = model.predict(dvalid)
    score = log_loss(y_valid, y_pred)
    scores.append(score)
    print(f'logloss: {score:.4f}\n')
    
    fscore = model.get_score(importance_type = 'total_gain')
    fscore = sorted([(k,v) for k, v in fscore.items()], key=lambda tpl: tpl[1], reverse = True)
    fscore_dict = dict(fscore)
    fscore_dicts.append(fscore_dict)
    print('xgboost importance')
    print(fscore[:])

score_CV = np.mean(scores)
print(score_CV)
```

```python
fig = make_subplots(rows = 2,cols = 2,
                    subplot_titles=("fold1", "fold2", "fold3", "fold4"))
tuples = [(1,1),(1,2),(2,1),(2,2)]
for i in range(len(evals_results)):
    train_log = evals_results[i]['train']['logloss'] 
    eval_log = evals_results[i]['eval']['logloss'] 
    x = np.arange(1,1+len(train_log))
    fig.add_trace(go.Scatter(x = x, y = train_log,mode = 'lines',name = 'train'),
                 row = tuples[i][0],col = tuples[i][1])
    fig.add_trace(go.Scatter(x = x, y = eval_log,mode = 'lines',name = 'eval'),
                 row = tuples[i][0],col = tuples[i][1])
fig.update_layout(
    title = "fold1, fold2, fold3, fold4"
)
```
